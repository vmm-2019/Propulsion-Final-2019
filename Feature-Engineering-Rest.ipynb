```python
df = df_master.copy()
```

```python
def insert_difference_col(df1, col_name, lag=1):
    df = df1.copy()
    #print(df.shape)
    new_col_name = col_name + '_delta'
    if lag > 1:
        new_col_name += '_' + str(lag)
    df[new_col_name] = df[col_name].diff(periods=lag)
    #print(df.shape, type(df[new_col_name]), np.array(df[new_col_name]))
    #print(df.columns[len(df.columns) -1])
    return df
```

```python
#df.drop(columns=['SHA_Wind_max_delta'], inplace=True)
df.shape
```

```python
cols_wsm = [c for c in df.columns if re.search('^[A-Z]{3}_Wind_max$', c)]
for c in cols_wsm:
    print(df.shape)
    df = insert_difference_col(df, c, lag=const.steps_30_mins)
    print(df.shape)
    df = insert_difference_col(df, c, lag=const.steps_1_hour)
    print(df.shape)
```

```python
cols_hum = [c for c in df.columns if re.search('^[A-Z]{3}_Hum$', c)]
```

```python
for c in cols_hum:
    df = insert_difference_col(df, c, lag=const.steps_30_mins)
    df = insert_difference_col(df, c, lag=const.steps_1_hour)
```

```python
cols_air = [c for c in df.columns if re.search('^[A-Z]{3}_Air_temp$]', c)]
for c in cols_air:
    df = insert_difference_col(df, c, lag=const.steps_30_mins)
    df = insert_difference_col(df, c, lag=const.steps_1_hour)
```

```python
cols_air = [c for c in df.columns if re.search('^[A-Z]{3}_Wind_speed$', c)]
for c in cols_air:
    df = insert_difference_col(df, c, lag=const.steps_30_mins)
    df = insert_difference_col(df, c, lag=const.steps_1_hour)
```

```python
cols_qnh = [c for c in df.columns if re.search('^QNH_', c)]
cols_qnh.sort()
cols_klo = [c for c in df.columns if re.search('^KLO_', c)]
cols_klo.sort()
cols_other = [c for c in df.columns if not(c in cols_qnh) and not(c in cols_klo) and c != 'datetime']
cols_other.sort()
cols_qnh_klo = [c for c in cols_qnh if re.search('KLO', c)]
cols_qnh_rest = [c for c in cols_qnh if not(c in cols_qnh_klo)]
#print(cols_qnh_rest)
cols_sorted = ['datetime'] + cols_klo + cols_other + cols_qnh_klo + cols_qnh_rest
cols_sorted
```

```python
df1 = df.copy()
df = df.reindex(columns=cols_sorted)
```
